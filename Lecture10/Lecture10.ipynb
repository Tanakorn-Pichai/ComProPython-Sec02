NUMPY MODULE

```python
import numpy as np

random_matrix =  np.random.randint(1, 11, size=(3, 3))
print("Random 3*3",random_matrix)  

matrix_sum = np.sum(random_matrix)
print("Sum",matrix_sum)

matrix_mean = np.mean(random_matrix)
print("Mean",matrix_mean)

transposed_matrix = np.transpose(random_matrix)
print("Transposed",transposed_matrix)
```

MATPLOTLIB MODULE

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

fig ,  ax = plt.subplots()

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

line , = ax.plot(x,y)

def update(frame):
    line.set_ydata(np.sin(x+frame /10.0))
    return line,
ani = animation.FuncAnimation(fig,update,frames=100,blit=True,interval=50)
plt.show()
```
